```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar la data cruda para ver el "estado inicial"
df = pd.read_csv('../data/raw/bank_churn.csv')
print(f"Total de registros a auditar: {len(df)}")
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Exited', data=df, palette='viridis')
plt.title('Distribución de Fuga de Clientes (0=Se queda, 1=Se va)')
plt.show()
```

```python
# Seleccionamos solo columnas numéricas para la correlación
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación de Variables Financieras')
plt.show()
```
